```python
def run_experiment( X, y, learning_rate = 0.1, epochs = 20 ):

  print(f"\n--- Running Experiment ---")
  model = Perceptron(n_features=X.shape[1], learning_rate=learning_rate)
  model.train(X, y, epochs)

  print("✅ Treinamento concluído!")
  print("Pesos finais:", model.w)
  print("Bias final:", model.b)
  Visualizer.plot_decision_boundary(X, y, model)

  return model
```

```python
#Atividades
#a) execute-o para o dataset da porta l´ogica OR (exemplo dos slides de aula);
#b) execute-o para o dataset da porta l´ogica AND;
#c) execute-o para o dataset da porta l´ogica XOR;

datasets = [ "or.csv", "and.csv", "xor.csv"]
columns_mask = [0, 1, 1, 1]
for ds in datasets:
  file_path = f"{PATH}/data-sets/{ds}"
  X, y = DataManager.prepare_dataset(file_path, [0,1,1,1], "D", 1)
  run_experiment(X, y, learning_rate=0.1, epochs=30)
```

```python
#Atividade
"""
d) Execute-o para o dataset Iris, reduzindo-o para um problema de classifica¸cão binária (classe 1 = setosa
vs. classe 2 = virginica + versicolor). O modelo pode ser treinado com os quatro atributos originais ou
com apenas dois de livre escolha, que facam a separacão binária das classes (ver Figura 1).
"""

datasets = [ "Iris.csv", "Iris.csv", "Iris.csv"]
columns_to_read_mask = [[0, 1, 1, 1, 1, 1], [0, 1, 1, 0, 0, 1], [0, 0, 1, 0, 1, 1] ]
label_column = "Species"
positive_label = "Iris-setosa"

for i, ds in enumerate(datasets):
  file_path = f"{PATH}/data-sets/{ds}"
  X, y = DataManager.prepare_dataset(file_path, columns_to_read_mask[i], label_column, positive_label)
  run_experiment(X, y, learning_rate=0.1, epochs=100)
```
